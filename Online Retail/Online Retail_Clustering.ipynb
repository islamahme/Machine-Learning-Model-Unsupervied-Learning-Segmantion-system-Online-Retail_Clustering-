# RFM Analysis

```python
import datetime as dt
```

```python
df['InvoiceDate'].max()
```

```python
Today_ddate=dt.datetime(2011,12,11)
```

```python
from datetime import datetime

# Assuming Today_date represents the current date
Today_date = datetime.now()

# Define the aggregation using groupby and agg
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda date: (Today_ddate - date.max()).days,
    'InvoiceNo': lambda num: num.nunique(),
    "Total_Price": lambda TotaPrice: TotaPrice.sum()
})
rfm
```

### The process of converting column names to["Recency","Frequency","Monetary"]

```python
rfm.columns=["Recency","Frequency","Monetary"]
rfm
```

```python
rfm.describe().T
```

We will reverse the days, meaning that it is logical that large days indicate that they are old, so we will start with the opposite

```python
rfm['Recency_score']=pd.qcut(rfm['Recency'],5,labels=[5,4,3,2,1])
```

```python
rfm['Frequency_score']=pd.qcut(rfm['Frequency'].rank(method='first'),5,labels=[1,2,3,4,5])
```

```python
rfm['Monetary_score']=pd.qcut(rfm['Monetary'],5,labels=[1,2,3,4,5])
```

```python
rfm.head()
```

```python
rfm["RFM_SCORE_CODE"]=(rfm['Recency_score'].astype(str)+rfm['Frequency_score'].astype(str))
rfm.head()
```

```python
Seg_map={
    r'[1-2][1-2]':'hibernating',
    r'[1-2][3-4]':'at_risk',
    r'[1-2]5':'can_loose',
    r'3[1-2]':'about_to_sleep',
    r'33':'need_attention',
    r'[3-4][4-5]':'loyal_customers',
    r'41':'promising',
    r'51':'new_customer',
    r'[4-5][2-3]':'potential_loyalists',
    r'5[4-5]':'champions'
    
}
rfm['Seggment']=rfm['RFM_SCORE_CODE'].replace(Seg_map,regex=True)
rfm
```

```python
rfm[["Seggment",'Recency','Frequency','Monetary']].groupby('Seggment').agg(["mean","count","max"]).round(2)
```
